# **1.Setup: Import libraries and dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your cleaned Fake News dataset
df = pd.read_csv('/FakeNewsDetection_Cleaned.csv')

# Show first few rows
df.head()
```

**Check basic info before plotting**

```python
print(df.shape)
print(df['subject'].value_counts())
print(df['date'].head())
```

# **2.Class Task_ Basics Plot**

a) Bar Chart — Count of Fake News by Subject

* You can see which subject (e.g., politics, world news, etc.)
has the highest number of fake articles.

```python
plt.figure(figsize=(8,5))
sns.countplot(x='subject', data=df, hue='subject', palette='coolwarm', legend=False)
plt.title('Number of Fake News Articles by Subject')
plt.xlabel('Subject')
plt.ylabel('Count')
plt.show()
```

b) Histogram — Distribution of Text Lengths

* Shows how long fake news articles usually are — most are around the same word length, but some are much longer or shorter.

```python
# Create a new column for text length if not done already
df['text_length'] = df['text'].apply(len)

plt.figure(figsize=(8,5))
sns.histplot(df['text_length'], bins=50, kde=True, color='skyblue')
plt.title('Distribution of Fake News Article Lengths')
plt.xlabel('Text Length')
plt.ylabel('Frequency')
plt.show()
```

c) Scatter Plot — Relationship between Text Length and Subject

*  Shows how article lengths differ across subjects — some topics tend to have longer fake news articles than others.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=range(len(df)), y='text_length', hue='subject', data=df, alpha=0.6)
plt.title('Text Length Variation by Subject')
plt.xlabel('Article Index')
plt.ylabel('Text Length')
plt.show()
```

## **Assignment3: Create 5 different Plots**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your cleaned Fake News dataset
df = pd.read_csv('/FakeNewsDetection_Cleaned.csv')

# (Optional) add text_length column again if not already saved
df['text_length'] = df['text'].apply(len)

# Quick check
df.head()

```

# **1. Boxplot – Text Length by Subject**

**Insight:**
Some subjects have a wider range of article lengths, meaning fake news stories in certain topics tend to be longer or shorter.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='subject', y='text_length', data=df, palette='Set2')
plt.title('Distribution of Text Lengths by Subject')
plt.xlabel('Subject')
plt.ylabel('Text Length')
plt.show()
```

# **2. Bar Plot – Articles by Subject**

```python
plt.figure(figsize=(8,5))
sns.countplot(x='subject', data=df, hue='subject', palette='coolwarm', legend=False)
plt.title('Number of Fake News Articles by Subject')
plt.xlabel('Subject')
plt.ylabel('Count')
plt.show()

```

# **3. Heatmap – Correlation Matrix**

**Insight:**
The heatmap shows correlations between numeric features (right now only text_length, but more features can be added later when model training starts).

```python
plt.figure(figsize=(4,3))
sns.heatmap(df[['text_length']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# **4. Pairplot – Relationship Exploration (Optional Advanced)**

**Insight:**
Pairplot visually confirms that text_length doesn’t have extreme outliers or clustering patterns.

```python
# Create a small sample for faster plotting
sample_df = df.sample(500, random_state=42)

sns.pairplot(sample_df[['text_length']], corner=True)
plt.suptitle('Pairplot: Relationship Between Features', y=1.02)
plt.show()
```

# **Plot 5: Top 20 Most Common Words in Fake News**

This chart shows which words appear most often in fake news articles.
Common terms like “breaking,” “report,” “says,” or “president” often occur, suggesting patterns or emotional hooks used in fake headlines.

```python
from collections import Counter
import re

# Combine all text into one big string
all_text = ' '.join(df['text'].astype(str))

# Remove punctuation and split into words
words = re.findall(r'\b\w+\b', all_text.lower())

# Count most common words (excluding stopwords)
common_words = Counter(words).most_common(20)

# Convert to DataFrame for plotting
word_df = pd.DataFrame(common_words, columns=['Word', 'Count'])

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x='Count', y='Word', data=word_df, palette='magma')
plt.title('Top 20 Most Common Words in Fake News Articles')
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.show()
```
