# **Step 1: Load Dataset**

```python
import pandas as pd

# Load your Fake News dataset
df = pd.read_csv('/Fake.csv')

# Show first 5 rows
df.head()
```

# **Step 2: Check Dataset Information**

```python
# Shape (rows & columns)
print("Shape:", df.shape)
```

```python
# Data types & missing values
print("\nInfo:")
df.info()
```

```python
# Summary stats
print("\nDescribe:")
df.describe(include='all')
```

# **Step 3: Check for Missing Values**

```python
# Missing values count
print("\nMissing values per column:")
df.isnull().sum().sort_values(ascending=False)
```

# **Step 4: Remove duplicates**

```python
Rduplicate = df.duplicated().sum()
print(f"Number of duplicate rows: {Rduplicate}")
```

# **Check Missing Values**

```python
missingVal = df.isnull().sum().sum()
print(f"Missing value in DATASET:{missingVal}")
```

```python
print(df.isnull().sum())
```

# **Step 5: Detect “Outlier-like” Text Lengths**

```python
# Check the length of text column
df['text_length'] = df['text'].apply(len)

print("Average text length:", df['text_length'].mean())
print("Maximum text length:", df['text_length'].max())
print("Minimum text length:", df['text_length'].min())

# Optional: identify very short texts (potential outliers)
short_texts = df[df['text_length'] < 50]
print("Short text articles:", len(short_texts))
```

# **Step 6: Remove Outlier-like Short Articles**

```python
# Remove short/incomplete text articles
df = df[df['text_length'] > 50]

# Check new shape
print("Shape after cleaning:", df.shape)
```
